```python
data = pd.read_csv('./hotel_bookings.csv')
data = data.dropna(subset=['country'])

import pycountry
def get_country_name(country_code):
    try:
        country = pycountry.countries.get(alpha_3=country_code)
        return country.name if country else country_code
    except KeyError:
        return country_code  


print(data.shape[0])
#data=data[data['hotel']=='City Hotel']
data_resort=data[data['hotel']=='Resort Hotel']
total_data=data_resort.shape[0]
data_resort=data_resort[data_resort['country']!='PRT']
print(data_resort.shape[0]/total_data)

data_resort['country']=data_resort['country'].apply(get_country_name)

data_city=data[data['hotel']=='City Hotel']
total_data=data_city.shape[0]
data_city=data_city[data_city['country']!='PRT']
print(data_city.shape[0]/total_data)

data_city['country']=data_city['country'].apply(get_country_name)
```
